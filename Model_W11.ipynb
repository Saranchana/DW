```
import pandas as pd
df =pd.read_csv ('https://raw.githubusercontent.com/Saranchana/DW/master/DIPEP.csv', header=None)
df.head()
```

```
df.isnull().any()
```

แยก Label ออกจาก Data

```
df_clean = df.dropna()
y =df_clean.iloc[:,0]
X =df_clean.iloc[:,1:401]
```

```
y.shape
```

```
X.shape
```

```

from sklearn.model_selection import train_test_split
X_train, X_val, y_train ,y_val= train_test_split(X, y, test_size=0.20, random_state=0)
```

```
X_train.shape
```

```
X_val.shape
```

```
from sklearn.preprocessing import MinMaxScaler
scl = MinMaxScaler()
scl.fit(X_train)
X_train_norm = scl.transform(X_train)
X_val_norm = scl.transform(X_val)
```

```
X_test_norm.max(axis=0)
```

```
X_val_norm.max(axis=0)
```

```
max(y_val)
```

#LogisticRegression

```
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression(C=12,intercept_scaling=0.2)
clf
```

# K-Nearwst Neighor

```
#Tune Model and Select model based on Training (CV) Accuracy
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=5, weights='uniform', algorithm='auto', leaf_size=30, p=2, metric='minkowski', metric_params=None, n_jobs=None)
from sklearn.model_selection import cross_val_score
accuracy = cross_val_score(clf, X_train_norm, y_train, cv=5)
accuracy.mean()
```

```
#Test Model besed on Validation Accuracy
clf.fit(X_train_norm,y_train)
y_predict = clf.predict(X_val_norm)
sum(y_predict == y_val) / len(y_val)
```

#Naive Bayes

```
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB(priors=None, var_smoothing=1e-01)
from sklearn.model_selection import cross_val_score
accuracy = cross_val_score(clf, X_train_norm, y_train, cv=5)
accuracy.mean()
```

```
#Test Model besed on Validation Accuracy
clf.fit(X_train_norm,y_train)
y_predict = clf.predict(X_val_norm)
sum(y_predict == y_val) / len(y_val)
```

#Decision Tree

```
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=0)
from sklearn.model_selection import cross_val_score
accuracy = cross_val_score(clf, X_train_norm, y_train, cv=5)
accuracy.mean()
```

```
#Test Model besed on Validation Accuracy
clf.fit(X_train_norm,y_train)
y_predict = clf.predict(X_val_norm)
sum(y_predict == y_val) / len(y_val)
```
