```
import pandas as pd
import numpy as np
```

#1. Data Gathering

```
df = pd.read_csv('https://raw.githubusercontent.com/plenoi/CMU_DataScience/master/DIPEP.csv')
df
```

#2. Data Pre-Processing

```
df.isnull().sum(axis=0)
```

แยก Label ออกจาก Data

```
df_clean = df.dropna()
y =df_clean.iloc[:,0]
X =df_clean.iloc[:,1:401]
```

```

y.shape
```

```
X.shape
```

```
from sklearn.model_selection import train_test_split
X_train, X_val, y_train ,y_val= train_test_split(X, y, test_size=0.20, random_state=0)
```

```
X_train.shape
```

```

X_val.shape
```

```
from sklearn.preprocessing import MinMaxScaler
scl = MinMaxScaler()
scl.fit(X_train)
X_train_norm = scl.transform(X_train)
X_val_norm = scl.transform(X_val)
```

```
X_val_norm.max(axis=0)
```

```
max(y_val)
```

#Support Vector Machine

```
from sklearn.svm import SVC
clf = SVC(C=2.0, degree=6, gamma='scale')
from sklearn.model_selection import cross_val_score
accuracy = cross_val_score(clf, X_train_norm, y_train, cv=5)
accuracy.mean()
```

```
#Test Model besed on Validation Accuracy
clf.fit(X_train_norm,y_train)
y_predict = clf.predict(X_val_norm)
sum(y_predict == y_val) / len(y_val)
```
