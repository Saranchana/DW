```
import pandas as pd
df =pd.read_csv ('https://raw.githubusercontent.com/YhokSiri/DataMining/master/DIPEP.csv', header=None)
df.head()
```

```
df_clean = df.dropna()
df_clean.shape
```

```
df_clean = df.dropna()
df_clean.shape
```

```
y = df_clean.iloc[:,0]
X = df_clean.iloc[:,1:401]
```

```
y.shape
```

```
X.shape
```

#Split

```
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.25, random_state=0)
```

```
X_train.shape
```

```
X_val.shape
```

```
from sklearn.preprocessing import MinMaxScaler
scl = MinMaxScaler()
scl.fit(X_train)
X_train_norm = scl.transform(X_train)
X_val_norm = scl.transform(X_val)
```

```
X_train_norm.max(axis=0)
```

```
X_val_norm.max(axis=0)
```

#LogisticRegression

```
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(C=1,intercept_scaling=0.25)
clf
```

```
clf.fit(X_train_norm,y_train)
```

```
y_predict = clf.predict(X_val_norm)
sum(y_predict == y_val) / len(y_val)
```
